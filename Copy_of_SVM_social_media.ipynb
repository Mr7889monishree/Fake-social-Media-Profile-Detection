```python
#separating data and labels , here label is the output which is 'fake' column
x_train=df_train.drop(columns='fake',axis=1) # drop remove output column
#for column axis=1 for row axis=0
x_train
```

```python
x_test=df_test.drop(columns='fake',axis=1)
x_test
```

```python
# Y variable stores the value of output column which is fake column
y_train=df_train['fake']
y_train
```

```python
y_test=df_test['fake']
y_test
```

#MODEL EVALUATION


```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

#  X_train, X_test, y_train, y_test are already defined

# Standardize the input features
scaler_x = StandardScaler()
X_train = scaler_x.fit_transform(x_train)
X_test = scaler_x.transform(x_test)
```

```python
# Print the shapes of training and testing datasets
print("X_train shape:", x_train.shape)
print("X_test shape:", x_test.shape)
print("y_train shape:", y_train.shape) # have only one column
print("y_test shape:", y_test.shape)
```

```python
# Initialize and train the SVM model
svm_model = SVC()
svm_model.fit(X_train, y_train)

```

```python
# Make predictions on the test set
y_prediction= svm_model.predict(X_test)
# Evaluate the model (optional)
accuracy = accuracy_score(y_test, y_prediction)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import precision_score, recall_score
```

```python
precision = precision_score(y_test, y_prediction, average='binary')
recall = recall_score(y_test, y_prediction, average='binary')

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
```

#predictive system

```python
# Input data for prediction (example values)
input_data = (0,0.62,1,0.4,0,0,0,0,1,17,34)

# Convert input data into a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# Reshape the array to match the expected input shape for the model (1 sample, multiple features)
input_data_reshape = input_data_as_numpy_array.reshape(1, -1)

# Standardize the input data using the same scaler
std_data = scaler_x.transform(input_data_reshape)

# Print standardized data
print("Standardized Data:", std_data)

# Make a prediction with the SVM model
prediction = svm_model.predict(std_data)

# Print the prediction
print("Prediction:", prediction)

# Interpret the result
if prediction[0] == 0:
    print('This is the true profile')
else:
    print('This is the fake profile')
```
